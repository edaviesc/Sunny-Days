```python
import matplotlib.pyplot as plt
sunny_days = [8, 10, 7, 14, 15, 20, 25, 19, 18, 14, 12, 7]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
avg_sunny_days = sum(sunny_days) / len(sunny_days)
plt.plot(months, sunny_days, marker="o", label="Sunny days")
plt.title("Sunny days")
plt.xlabel("Months")
plt.ylabel("Sunny days")
plt.axhline(y=avg_sunny_days, linestyle="--", color="orange", label="Average")
plt.legend()
plt.show()
```
